```python
df.drop(['Inside_Outside', 'Weapon','New_District'], axis=1, inplace=True)
```

```python
df.drop('Ethnicity',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# Drop rows with missing latitude and longitude values
df.dropna(subset=['Latitude', 'Longitude'], inplace=True)
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)
```

```python
categorical_columns = ['PremiseType', 'Neighborhood', 'Race', 'Post', 'Old_District']
df[categorical_columns] = df[categorical_columns].fillna('Unknown')
```

```python
df.isnull().sum()
```

```python
df = df[(df['Age'] >= 0) & (df['Age'] <= 100)]
```

```python
df = df[(df['Latitude'].between(39.2, 39.4)) & (df['Longitude'].between(-76.7, -76.5))]
```

```python
df['CrimeDateTime'] = pd.to_datetime(df['CrimeDateTime'], errors='coerce')
df['Hour'] = df['CrimeDateTime'].dt.hour
df['DayOfWeek'] = df['CrimeDateTime'].dt.dayofweek
df['Month'] = df['CrimeDateTime'].dt.month
```

```python
df
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.drop('Total_Incidents', axis=1, inplace=True)
```
